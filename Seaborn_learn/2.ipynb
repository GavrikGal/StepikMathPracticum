```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
%matplotlib inline
sns.set()
```

```python
df = sns.load_dataset('titanic')
df.head(5)
```

```python
sns.countplot(x='pclass', data=df, hue='sex')
```

```python
sns.countplot(x='deck', data=df, hue='sex')
```

```python
sns.countplot(x='deck', data=df, hue='class')
```

```python
fig = sns.FacetGrid(df, col='survived')
fig = fig.map(sns.countplot, 'deck', hue='class', data=df, palette='Set1')
```
