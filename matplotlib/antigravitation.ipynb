```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
with open('./img/1.png', 'rb') as map_file:
    map_img1 = plt.imread(map_file)

with open('./img/2.png', 'rb') as map_file:
    map_img2 = plt.imread(map_file)

with open('./img/3.png', 'rb') as map_file:
    map_img3 = plt.imread(map_file)

with open('./img/4.png', 'rb') as map_file:
    map_img4 = plt.imread(map_file)
```

```python
fig = plt.figure(figsize=(8, 9))
axes0 = fig.add_axes((0, 0, 1, 1))
axes1 = fig.add_axes((0, 0.5, 1, 0.5))
axes1.imshow(map_img1)
axes1.axis('off')
axes2 = fig.add_axes((0.05, 0.01, 0.3, 0.5))
axes2.imshow(map_img2)
axes2.axis('off')
axes3 = fig.add_axes((0.36, 0.01, 0.3, 0.5))
axes3.imshow(map_img3)
axes3.axis('off')
axes4 = fig.add_axes((0.67, 0.01, 0.3, 0.5))
axes4.imshow(map_img4)
axes4.axis('off')
axes0.axis('off')
plt.show()
```
