# Изучение функций с помощью MatplotLib
## Задача № 1

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

def f(x, a, b):
    return (x**b + a**b) / x**b

def make_plot(axes, results, x, args, ylim=(-10, 10),
              xlim=(-5, 5), title=False, legend=False, color=None):
    for result, ab in zip(results, args):
        axes.plot(x, result, label=f'$α = {ab[0]}$, $β = {ab[1]}$', color=color)
    axes.set_xlabel('$x$', fontsize=20)
    axes.set_ylabel('$f(x)$', fontsize=20)
    axes.set_ylim(*ylim)
    axes.set_xlim(*xlim)
    if title:
        axes.set_title('График функции $f(x) = \\frac{x^β + α^β}{x^β}$ при различных $α$ и $β$', fontsize=20)
    if legend:
        axes.legend(fontsize=18)
```

### Часть 1

```python
x = np.linspace(-5, 5, 101)

ab1 = (1, 1)
ab2 = (2, 1)
ab3 = (1, 2)

args = (ab1, ab2, ab3)

results = [f(x, a, b) for a, b in args]
```

```python
fig, axes = plt.subplots(figsize=(9, 9))

make_plot(axes, results, x, args, title=True, legend=True)

fig.show()
```

### Часть 2

```python
x = np.arange(0.01, 5, 0.01)
results = [f(x, a, b) for a, b in args]

x_small = np.geomspace(0.001, 1)
results_small = [f(x_small, a, b) for a, b in args]

x_big = np.geomspace(1, 10**3)
results_big = [f(x_big, a, b) for a, b in args]
```

```python
fig = plt.figure(figsize=(9, 9))

axes = fig.add_axes([0, 0, 1, 1])
make_plot(axes, results, x, args, ylim=(0, 10), xlim=(0, 5),
          title=True, legend=True)

axes_small = fig.add_axes([0.4, 0.6, 0.15, 0.35])
make_plot(axes_small, results_small, x_small, args,
          ylim=(0, 10**3), xlim=(0, 0.07))

axes_big = fig.add_axes([0.6, 0.4, 0.35, 0.15])
make_plot(axes_big, results_big, x_big, args,
          ylim=(1, 1.5), xlim=(0, 10**2))

fig.show()
# fig.savefig('1.svg')
```

### Часть 3

```python
x = np.arange(-0.01, -5, -0.01)
results = [f(x, a, b) for a, b in args]

x_small = np.geomspace(-1, -1e6)
results_small = [f(x_small, a, b) for a, b in args]
```

```python
fig = plt.figure(figsize=(9, 9))

axes = fig.add_axes([0, 0, 1, 1])
make_plot(axes, results, x, args, ylim=(-5, 5), xlim=(-5, 0),
          title=True, legend=True)

axes_small = fig.add_axes([0.1, 0.1, 0.5, 0.3])
make_plot(axes_small, results_small, x_small, args,
          ylim=(-0.1, 1.2), xlim=(-1e2, 0))
plt.grid(True)
```

### Выводы
Я молодец

## Задача 2

## Задача 3

```python
x = np.linspace(0.01, 10, 101)

ab1 = (1, 0.5)
ab2 = (1, -0.5)
ab3 = (1, -1.5)



args = (ab1, ab2, ab3)

results = [f(x, a, b) for a, b in args]

fig, axes = plt.subplots(figsize=(9, 9))

make_plot(axes, results, x, args, ylim=(0, 10), xlim=(0, 10),
          title=True, legend=True)

fig.show()
```

```python
x = np.linspace(0.01, 5, 101)

ab1 = (1, 0)
ab2 = (1, -1)

ab3 = (1, 0.5)
ab4 = (1, 0.8)

fig = plt.figure(figsize=(13, 4))

args1 = (ab1, ab2, ab3, ab4)
results1 = [f(x, a, b) for a, b in args1]

axes1 = fig.add_axes([0, 0, 0.27, 1])
make_plot(axes1, results1, x, args1, ylim=(0, 5), xlim=(0, 5), legend=True)

ab3 = (1, -0.5)
ab4 = (1, -0.8)

args2 = (ab1, ab2, ab3, ab4)
results2 = [f(x, a, b) for a, b in args2]

axes2 = fig.add_axes([0.33, 0, 0.27, 1])
make_plot(axes2, results2, x, args2, ylim=(0, 5), xlim=(0, 5), legend=True, title=True)

ab3 = (1, -1.5)
ab4 = (1, -2.5)

args3 = (ab1, ab2, ab3, ab4)
results3 = [f(x, a, b) for a, b in args3]

axes3 = fig.add_axes([0.66, 0, 0.27, 1])
make_plot(axes3, results3, x, args3, ylim=(0, 5), xlim=(0, 5), legend=True)

fig.tight_layout()
fig.show()
```
