## Задача №1

### Часть 1

Построить в общих осях графики для:

- $\alpha = 1, \beta = 1$

- $\alpha = 2, \beta = 1$

- $\alpha = 1, \beta = 2$

```python
x = np.linspace(-5,5,201)

def f(x):
    return (x**b+a**b)/x**b
```

```python
fig, axes = plt.subplots(figsize=(16, 12))

a, b = 1, 1
axes.plot(x, [f(x) for x in x], 'r', label=r'$\alpha$ = 1, $\beta$ = 1')

a, b = 2, 1
axes.plot(x, [f(x) for x in x], 'b', label=r'$\alpha$ = 2, $\beta$ = 1')

a, b = 1, 2
axes.plot(x, [f(x) for x in x], 'g', label=r'$\alpha$ = 1, $\beta$ = 2')

axes.legend(fontsize=14)
axes.set_xlabel('X', fontsize=16)
axes.set_ylabel('f(X)', fontsize=16)
axes.set_title(
    r'График функции f(x)=$\frac{x^{\beta}+\alpha^{\beta}}{x^{\beta}}$ при различных $\alpha$ и $\beta$', fontsize=16)
axes.set_ylim(-20, 30)

plt.savefig('1.svg')
```

## Задача №3

Построить в общих осях графики для:

- $\alpha = 1, \beta = 0.5$

- $\alpha = 1, \beta = -0.5$

- $\alpha = 1, \beta = -1.5$

Сделайте выводы о поведении графиков, включая возрастание/убывание и выпуклость/вогнутость

```python
x = np.linspace(-50,50,1001)

a = [1, 1, 1]
b = [0.5, -0.5, -1.5]

fig = plt.figure(figsize=(14, 10))
axes = fig.add_axes([0, 0, 1, 1])

for i, j in zip(a, b):
    a, b = i, j
    axes.plot(x, [f(x) for x in x], label=fr'$\alpha$={i}, $\beta$={j}')

axes.set_ylim(0,100)
axes.set_xlim(-1,40)
axes.legend(fontsize=16)
axes.set_xlabel('X', fontsize=12)
axes.set_ylabel('f(X)', fontsize=12)
axes.set_title(
    r'График функции f(x)=$\frac{x^{\beta}+\alpha^{\beta}}{x^{\beta}}$ при различных $\alpha$ и $\beta$', fontsize=18)

a = [1, 1, 1]
b = [0.5, -0.5, -1.5]

axes = fig.add_axes([0.5, 0.5, 0.4, 0.3])
for i, j in zip(a, b):
    a, b = i, j
    axes.plot(x, [f(x) for x in x])
    
axes.set_ylim(0,6)
axes.set_xlim(-1,6)
axes.set_xlabel('X', fontsize=12)
axes.set_ylabel('f(X)', fontsize=12)
axes.set_title(
    r'График функции f(x) при X$\subset$(-1, 6)', fontsize=18)

plt.savefig('5.svg')
```

1. Все графики отображаются только при положительных значениями X. При отрицательных Х значений функции не существует.

2. График функции при положительном $\beta$=0.5 убывающий, остальные при отрицательных $\beta$ - возрастающие.

3. Графики функций пересекаются в точке (1, 2).

4. График функции при $\beta$=0.5 и -1.5 вогнутый, при $\beta$=-0.5 выпуклый.
