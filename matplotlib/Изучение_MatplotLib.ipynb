## Задача №1

### Часть 1

Построить в общих осях графики для:

- $\alpha = 1, \beta = 1$

- $\alpha = 2, \beta = 1$

- $\alpha = 1, \beta = 2$

```python
def f(x, alpha, beta):
     with np.errstate(divide='ignore', invalid='ignore'):
            return (x**beta+alpha**beta)/(x**beta)

x=np.linspace(-5,5,201)

fig, axes= plt.subplots(figsize=(13,8))
axes.plot(x, f(x, 1, 1), 'r', label='$\\alpha=1, \\beta=1$')
axes.plot(x, f(x, 2, 1), 'b', label='$\\alpha=2, \\beta=1$')
axes.plot(x, f(x, 1, 2), 'g', label='$\\alpha=1, \\beta=2$')
axes.plot(x, [0 for i in x], 'k--')
axes.set_xlabel('x', fontsize=18)
axes.set_ylabel('y', fontsize=18)
axes.set_ylim([-20,30])
axes.set_title(r'График функции $f(x)=\frac{x^\beta+\alpha^\beta}{x^\beta}$ при разных значениях $\alpha$ и $\beta$', fontsize=18)
axes.legend(loc=0, fontsize=18)

plt.savefig('C:\\Users\\123\\OneDrive\\Desktop\\py4e\\data\\1.svg')
plt.savefig('C:\\Users\\123\\OneDrive\\Desktop\\py4e\\data\\1.jpg')
```

## Задача №3

Построить в общих осях графики для:

- $\alpha = 1, \beta = 0.5$

- $\alpha = 1, \beta = -0.5$

- $\alpha = 1, \beta = -1.5$

Сделайте выводы о поведении графиков, включая возрастание/убывание и выпуклость/вогнутость

```python
x=np.linspace(0,5,201)

fig, axes= plt.subplots(figsize=(13,8))
axes.plot(x, f(x, 1, 0.5), label='$\\alpha=1, \\beta=0.5$')
axes.plot(x, f(x, 1, -0.5), label='$\\alpha=1, \\beta=-0.5$')
axes.plot(x, f(x, 1, -1.5), label='$\\alpha=1, \\beta=-1.5$')
axes.plot(x, [0 for i in x], 'k--')
axes.set_xlabel('x', fontsize=18)
axes.set_ylabel('y', fontsize=18)
#axes.set_ylim([-10,10])
axes.set_title(r'График функции $f(x)=\frac{x^\beta+\alpha^\beta}{x^\beta}$ при разных значениях $\alpha$ и $\beta$', fontsize=18)
axes.legend(loc=0, fontsize=18)

def g(x, alpha, beta):
     with np.errstate(divide='ignore', invalid='ignore'):
            return -(alpha**beta)*beta*(-1-beta)*(x**(-2-beta))

        
fig, axes= plt.subplots(figsize=(13,8))
axes.plot(x, g(x, 1, 0.5), label='$\\alpha=1, \\beta=0.5$')
axes.plot(x, g(x, 1, -0.5), label='$\\alpha=1, \\beta=-0.5$')
axes.plot(x, g(x, 1, -1.5), label='$\\alpha=1, \\beta=-1.5$')
axes.plot(x, [0 for i in x], 'k--')
axes.set_xlabel('x', fontsize=18)
axes.set_ylabel('y', fontsize=18)
axes.set_ylim([-10,10])
axes.set_title(r'График функции $g(x) (вторая производная)$ при разных значениях $\alpha$ и $\beta$', fontsize=18)
axes.legend(loc=0, fontsize=18)

plt.savefig('C:\\Users\\123\\OneDrive\\Desktop\\py4e\\data\\5.svg')
plt.savefig('C:\\Users\\123\\OneDrive\\Desktop\\py4e\\data\\5.jpg')
```
