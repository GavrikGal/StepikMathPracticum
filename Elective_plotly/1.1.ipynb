```python
import plotly
import plotly.graph_objs as go
import plotly.express as px
from plotly.subplots import make_subplots

import numpy as np
import pandas as pd
```

```python
def f(x):
    return np.sin(x)

def g(x):
    return 1 - np.cos(x)
```

```python
x1 = np.arange(-2 * np.pi, 2 * np.pi, 0.01*np.pi)
x2 = [-2*np.pi, -3*np.pi/2, 0, np.pi/2, 2*np.pi]
```

```python
fig = go.Figure()

fig.update_xaxes(zeroline=True, zerolinewidth=3)
fig.update_yaxes(zeroline=True, zerolinewidth=3)

fig.add_trace(go.Scatter(x=x1, y=f(x1), name='f(x)=sin(x)'))
fig.add_trace(go.Scatter(x=x1, y=g(x1), name='g(x)=1-cos(x)'))
fig.add_trace(go.Scatter(x=x2, y=f(x2), name='Точки пересечения', mode='markers'))

fig.update_layout(legend_orientation='h',
                  legend=dict(x=.5, xanchor='center'),
                  title='Задание 1',
                  xaxis_title='x',
                  yaxis_title='f(x), g(x)',
                  margin=dict(l=5, r=5, t=35, b=5))

fig.show()
```
